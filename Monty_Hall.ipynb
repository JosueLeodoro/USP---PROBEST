# O problema de Monty Hall:

```python
import numpy as np
import random
# define a semente do gerador de números aleatórios
np.random.seed(41)
muda_porta = True # decide se muda a porta
if(muda_porta == True):
    print('Opção: Muda a porta')
else:
    print('Opção: Não muda a porta')
print('*************')
ns = 10 # número de simulações
vitoria = 0 # número de vezes que ganha o carro
derrota = 0 # número de derrotas
for s in range(0,ns):
    print('Round:', s+1)
    portas = [1,2,3]
    # coloca o carro em uma porta
    porta_car = random.choice(portas)
    # seleciona uma porta
    porta_sel = random.choice(portas)
    print('Porta selecionada:', porta_sel)
    print('Porta com o carro:', porta_car)
    # se o carro não está na porta selecionada
    if(porta_sel != porta_car):
        portas.remove(porta_sel)
        portas.remove(porta_car)
        porta_open = portas.pop()
        print('Porta aberta:', porta_open)
        # se mudar, vence
        if(muda_porta == True):
            vitoria = vitoria + 1
            print('Muda a porta: vitória')
        # se não mudar, perde o carro
        else:
            derrota = derrota + 1
            print('Não muda a porta: derrota')
    else:
        # se a porta selecinada contém o carro
        portas.remove(porta_sel)
        porta_open = random.choice(portas)
        print('Porta aberta:', porta_open)
        # se mudar, perde o carro
        if(muda_porta == True):
            derrota = derrota + 1
            print('Muda a porta: derrota')
        # se não mudar, vence
        else:
            vitoria = vitoria + 1
            print('Não muda a porta: vitória')
    print('*************')
print('Fração de vitórias:', vitoria/ns)
print('Fração de derrotas:', derrota/ns)
```
